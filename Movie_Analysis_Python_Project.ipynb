# Q4. *Time-Based Analysis*

## 1. How has the number of movies released per year changed over time? (Line chart).

```python
print(df.columns)
```

```python
df['ReleaseYear'].value_counts().sort_index().plot(kind='line')
plt.title("Movies Released Per Year")
plt.show()
```

## 2. Has the average movie budget increased over the decades? (Line/Bar chart).

```python
df['Decade'] = (df['ReleaseYear']//10)*10
df.groupby('Decade')['BudgetUSD'].mean().plot(kind='line',color="red")
plt.title("Average Budget by Decade")
plt.show()
```

## 3.  Do IMDb ratings show any trend across release years? (Line chart).

```python
df.groupby('ReleaseYear')['IMDbRating'].mean().plot(kind='line',color="green")
plt.title("IMDb Ratings Over Years")
plt.show()
```

## 4.  Which genre has grown most in popularity since 2000? (Line/Bar chart).

```python
df[df['ReleaseYear']>=2000]['Genre'].value_counts().head(10).plot(kind='bar',color="orange")
plt.title("Genre Popularity Since 2000")
plt.show()
```

## 5.  Compare average Global Box Office revenue of movies before 2000 vs after 2000 (Bar chart).

```python
df['period'] = df['ReleaseYear'].apply(lambda x: 'Before 2000' if x<2000 else 'After 2000')
df.groupby('period')['Global_BoxOfficeUSD'].mean().plot(kind='bar',color=['lightcoral','lightgreen'])
plt.title("Global Revenue Before vs After 2000")
plt.ylabel("Average Global Box Office (USD)")
plt.show()
```

# Q5. *Categorical Insights*

## 1.  Which director has the highest average IMDb rating? (Top 10 directors bar chart).

```python
df.groupby('Director')['IMDbRating'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='pink')
plt.title("Top 10 Directors by IMDb Rating")
plt.show()
```

## 2.  Which lead actors have appeared in the most movies? (Bar chart).

```python
df['LeadActor'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Lead Actors by Number of Movies")
plt.show()
```

## 3.  Which country produces the most movies? (Bar chart).

```python
df['Country'].value_counts().head(10).plot(kind='bar', color='green')
plt.title("Top 10 Movie Producing Countries")
plt.show()
```

## 4.  Which genre tends to have the highest US vs Global box office ratio?

```python
df['US_Global_Ratio'] = df['US_BoxOfficeUSD'] / df['Global_BoxOfficeUSD']
df.groupby('Genre')['US_Global_Ratio'].mean().sort_values(ascending=False).plot(kind='bar', color='teal')
plt.title("US vs Global Box Office Ratio by Genre")
plt.show()
```

## 5.  Compare IMDb and Rotten Tomatoes ratings for each genre — are they aligned? (Side-by-side bar chart)

```python
ratings = df.groupby('Genre')[['IMDbRating','RottenTomatoesScore']].mean()
ratings.plot(kind='bar')
plt.title("IMDb vs Rotten Tomatoes by Genre")
plt.show()
```

# Q6. *Advanced / Critical Thinking*

## 1.  Do higher IMDb ratings always mean higher box office revenue? (Correlation test).

```python
print("Correlation between IMDb Rating and Global Box Office:", df['IMDbRating'].corr(df['Global_BoxOfficeUSD']))
```

## 2. What percentage of movies had a budget above $100M but failed to make a profit (Global revenue < Budge 

```python
loss_movies = df[(df['BudgetUSD'] > 100_000_000) & (df['Global_BoxOfficeUSD'] < df['BudgetUSD'])]
percent_loss = len(loss_movies) / len(df) * 100
print(f"Movies with Budget > $100M but failed to profit: {percent_loss:.2f}%")
```

## 3. Which genres are most 'profitable' (average Global Box Office / Budget ratio)? (Bar chart).

```python
df['Profit_Ratio'] = df['Global_BoxOfficeUSD'] / df['BudgetUSD']
df.groupby('Genre')['Profit_Ratio'].mean().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title("Most Profitable Genres")
plt.show()
```

## 4.  Does a higher Rotten Tomatoes score guarantee higher first-week sales? (Scatter plot).

```python
sns.scatterplot(x='RottenTomatoesScore', y='One_Week_SalesUSD', data=df,color="red")
plt.title("Rotten Tomatoes Score vs First Week Sales")
plt.show()
```

## 5. Predictive thinking: Based on budget, genre, and director, what factors seem most important for a movie’s success?

```python
sns.heatmap(df[['BudgetUSD','US_BoxOfficeUSD','Global_BoxOfficeUSD','IMDbRating','RottenTomatoesScore']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Q7. *Visualization Ideas*

## 1. Heatmap of correlations among numeric variables (Budget, Revenue, Ratings, Votes).

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis")
plt.title("Numeric Feature Correlations")
plt.show()
```

## 2. Bubble chart of Budget vs Global Revenue, with bubble size = IMDb votes, color = Genre.

```python
plt.scatter(df['BudgetUSD'], df['Global_BoxOfficeUSD'], 
            s=df['NumVotesIMDb']/1000, 
            c=df['Genre'].astype('category').cat.codes, alpha=0.5)
plt.title("Budget vs Global Revenue (Bubble Size = IMDb Votes)")
plt.xlabel("Budget (USD)")
plt.ylabel("Global Revenue (USD)")
plt.show()
```

## 3. Trend line of IMDb rating vs Release Year.

```python
sns.regplot(x='ReleaseYear', y='IMDbRating', data=df, scatter=False, lowess=True, color='red')
plt.title("IMDb Rating Trend Over the Years")
plt.show()
```

## 4. Pie chart or Treemap of genre distribution.

```python
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Genre Distribution")
plt.ylabel("")
plt.show()
```

## 5. Box plot of IMDb rating by Genre.

```python
sns.boxplot(x='Genre', y='IMDbRating', data=df)
plt.xticks(rotation=90)
plt.title("IMDb Ratings by Genre")
plt.show()
```
